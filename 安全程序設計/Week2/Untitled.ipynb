```python
import requests
from bs4 import BeautifulSoup
r = requests.Session()
while True:
    index_html = r.post("http://140.134.25.138:10022/index.php")
    soup = BeautifulSoup(index_html.text)
    num1 = int(soup.find('span',{'id':'num1'}).text)
    num2 = int(soup.find('span',{'id':'num2'}).text)
    op = soup.find('span', {'id':'flag'}).text
    if(op == '+'):
        token = num1+num2
    elif(op == '-'):
        token = num1-num2
    elif(op == '*'):
        token = num1*num2
    #print(token)
    data = {'token':token}
    respone = r.post('http://140.134.25.138:10022/index.php',data=data)
    respone = BeautifulSoup(respone.text)
    message = respone.find('h1', {'name':'flag'}).text
    if(message != 'CTF{NOT_THIS_ONE}'):
        print(message)
        break
```
